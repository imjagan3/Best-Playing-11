```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline

pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
pd.set_option('display.width',None)

df=pd.read_csv('C:/Users/ELCOT/Desktop/Jagan/fifa20_data.csv')
df.head(10)
```

```python
a=0.5
b=1
c=2
d=3
```

```python
df.columns = df.columns.str.replace(' ','_').str.lower()
df.head(10)
```

```python

df['gk_stopper']=(d*df.gk_diving + d*df.gk_handling + c*df.gk_positioning + d*df.gk_reflexes + c*df.composure + b*df.vision + a*df.aggression +b*df.mentality+d*df.jumping + a*df.stamina)/(4*d +2*c+2*b+2*a) 
```

```python

df['gk_sweeper'] = (a*df.ball_control+d*df.acceleration+d*df.sprint_speed+b*df.interceptions+d*df.gk_positioning+d*df.gk_reflexes+d*df.gk_kicking)/(a+b+5*d) 
```

```python
plt.figure(figsize=(15,3))

viz1 = df.sort_values('gk_stopper',ascending=False)[:5]
x1 =np.array(list(viz1['name']))
y1  =np.array(list(viz1['gk_stopper']))

sb.barplot(x1,y1,palette='colorblind')
plt.ylabel('Shot_stopper')

```
